```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Importing the Image

**Greyscaled the Image, and displayed it**

```python
image = cv2.imread("Einstein.jpg")
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grey_im = np.array(grey_img)
plt.gray()
plt.imshow(grey_im)
plt.show()
```

# Part (a)

*Computing Forward, Backward and Central Gradient on the Image*

```python
print(grey_im.shape)
```

# Calculating the Forward Gradient

*Forward Gradient in X-Direction*

```python
#Forward Filter
fil = [1, -1]
f_f_image_x = np.zeros(shape = (2566, 2000))
for i in range(0, 2566):
    for j in range(0,1999):
        f_f_image_x[i][j] = grey_im[i][j] * fil[0] + grey_im[i][j + 1] * fil[1]
```

```python
plt.imshow(f_f_image_x)
plt.title("Forward Gradient in X - Direction")
plt.show()
```

*Forward Gradient in Y-Direction*

```python
filter_y = np.array([1, -1])
ff_image_y = np.zeros((2566, 2000))
for i in range(0, 2000):
    for j in range(0, 2565):
        ff_image_y[j][i] = grey_im[j][i] * filter_y[0] + grey_im[j + 1][i] * filter_y[1]
        
```

```python
plt.imshow(ff_image_y)
plt.title("Forward Gradient in Y-Direction")
plt.show()
```

# Calculating Gradient Magnitude for Forward Gradient

```python
grad_mag = np.zeros((2566, 2000))
for i in range(0, 2566):
    for j in range(0, 2000):
        grad_mag[i][j] = ((ff_image_y[i][j] ** 2) +  (f_f_image_x[i][j] ** 2)) ** 0.5
```

```python
grad_mag
```

```python
plt.imshow(grad_mag)
plt.gray()
plt.title("Gradient Magnitude")
plt.show()
```

# Calculating Gradient Direction

```python
import math
grad_dir = np.zeros((2566, 2000))
for i in range(0, 2566):
    for j in range(0, 2000):
        grad_dir[i][j] = math.atan(f_f_image_x[i][j] / ff_image_y[i][j])
        # grad_mag[i][j] = ((ff_image_y[i][j] ** 2) +  f_f_image[i][j] ** 2) ** 0.5
```

```python
grad_dir
```

```python
plt.imshow(grad_dir)
plt.title("Gradient Direction")
plt.show()
```

# Calculating the Central Gradient

*Central Gradient in X-Direction*

```python
# Central Filter
fil_c = [-1, 0, 1]
cf_image_x = np.zeros(shape = (2566, 2000))
for i in range(0, 2566):
    for j in range(1, 1999):
        cf_image_x[i][j] = grey_im[i][j - 1] * fil_c[0] + grey_im[i][j + 1] * fil_c[2]
```

```python
plt.imshow(cf_image_x)
plt.title("Central Gradient in X - Direction")
plt.show()
```

*Central Gradient in Y - Direction*

```python
cf_image_y = np.zeros(shape = (2566, 2000))
for i in range(0, 2000):
    for j in range(1, 2565):
        cf_image_y[j][i] = grey_im[j ][i] * fil_c[0] + grey_im[j + 1][i] * fil_c[2]
```

```python
plt.imshow(cf_image_y)
plt.title("Central Gradient in Y-Direction")
plt.show()
```

# Calculating Gradient Magnitude for Central Gradient

```python
grad_mag_central = np.zeros((2566, 2000))
for i in range(0, 2566):
    for j in range(0, 2000):
        grad_mag_central[i][j] = ((cf_image_x[i][j] ** 2) + (cf_image_y[i][j] ** 2))
```

```python
plt.imshow(grad_mag_central)
plt.title("Gradient Magnitude for Central Gradient")
plt.show()
```

# Calculating Gradient Direction for Central Gradient

```python
cc_grad_dir = np.zeros((2566, 2000))
for i in range(0, 2566):
    for j in range(0, 2000):
        cc_grad_dir[i][j] = math.atan(cf_image_x[i][j] / cf_image_y[i][j])
```

```python
plt.imshow(cc_grad_dir)
plt.title("Gradient Direction for Central Direction")
plt.show()
```
